```python
 import matplotlib.pyplot as plt
 import seaborn as sns
 import numpy as np
 import pandas as pd
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
```

## Matplotlib Line Plot

```python
 plt.figure(figsize=(6,4))
 plt.plot(x, y, color='blue', linestyle='-', label='Sine Wave')
 plt.xlabel("X Axis")
 plt.ylabel("Y Axis")
 plt.title("Matplotlib Line Plot")
 plt.legend()
 plt.show()
```

## Seaborn Line Plot

```python
 sns.set(style="darkgrid")
 sns.lineplot(x=x, y=y, color='red')
 plt.title("Seaborn Line Plot")
 plt.show()
```

##  Bar Chart (Comparison & Categories)


```python
categories = ["A", "B", "C", "D"]
values = [10, 20, 15, 25]
```

## Matplotlib Bar Chart

```python
 plt.figure(figsize=(6,4))
 plt.bar(categories, values, color=['red', 'blue', 'green', 
'purple'])
 plt.xlabel("Categories")
 plt.ylabel("Values")
 plt.title("Matplotlib Bar Chart")
 plt.show()
```

## Seaborn Bar Chart

```python
sns.barplot(x=categories, y=values, hue=categories, palette="viridis", legend=False)
plt.title("Seaborn Bar Chart")
plt.show()
```

##  Histogram (Distribution & Patterns)

```python
data = np.random.randn(1000)
```

## Matplotlib Histogram

```python
 plt.figure(figsize=(6,4))
 plt.hist(data, bins=30, color='blue', edgecolor='black', 
alpha=0.7)
 plt.xlabel("Value")
 plt.ylabel("Frequency")
 plt.title("Matplotlib Histogram")
 plt.show()
```

 ## Seaborn Histogram

```python
 sns.histplot(data, bins=30, kde=True, color='red')
 plt.title("Seaborn Histogram")
 plt.show()
```

##  Scatter Plot (Relationships & Correlations)

```python
 np.random.seed(42)
 x = np.random.rand(50)
 y = np.random.rand(50)
```

##  Matplotlib Scatter Plot


```python
 plt.figure(figsize=(6,4))
 plt.scatter(x, y, color='blue', alpha=0.5)
 plt.xlabel("X Axis")
 plt.ylabel("Y Axis")
 plt.title("Matplotlib Scatter Plot")
 plt.show()
```

##  Seaborn Scatter Plot

```python
 sns.scatterplot(x=x, y=y, color='red')
 plt.title("Seaborn Scatter Plot")
 plt.show()
```

## Heatmap (Correlation & Relations)

```python
np.random.seed(33)
data = np.random.rand(5, 5)
df = pd.DataFrame(data, columns=["A", "B", "C", "D", "E"])
```

## Seaborn Heatmap

```python
 sns.heatmap(df, annot=True, cmap="coolwarm")
 plt.title("Seaborn Heatmap")
 plt.show()
```
